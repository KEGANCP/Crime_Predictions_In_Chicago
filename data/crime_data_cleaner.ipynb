```python
#load imports
import pandas as pd
import datetime as dt
```

```python
#Load CSV file into df
crime_df = pd.read_csv('C:/Users/auste/Documents/crime_data_2001_present.csv')
crime_df.head()
```

```python
#Drop data prior to 2010
sample_crime_df = crime_df[crime_df['Year'] >= 2019]
sample_crime_df.rename(columns ={'Date': 'Datetime'}, inplace=True)

sample_crime_df
```

```python
#Split Date column into seperate Date and time columns
sample_crime_df['Date'] = pd.to_datetime(sample_crime_df['Datetime']).dt.date
sample_crime_df['Time'] = pd.to_datetime(sample_crime_df['Datetime']).dt.time
sample_crime_df
```

```python
#drop unneeded columns
sample_crime_df = sample_crime_df.drop(columns=['Case Number', 'Datetime', 'Updated On', 'X Coordinate', 'Y Coordinate', 'Location', 'Beat'])
sample_crime_df.head()
```

```python
sample_crime_df['Primary Type'].unique()
```

```python
#create violent crime column
violent_crimes = ['HOMICIDE', 'BATTERY', 'ASSAULT', 'CRIMINAL SEXUAL ASSAULT', 'ROBBERY']
violent = []

for row in sample_crime_df['Primary Type']:
    if row in violent_crimes:
        violent.append('VIOLENT')
    else:
        violent.append('NON-VIOLENT')
        
sample_crime_df.insert(4,'Violence Status',violent,True)
sample_crime_df.head()
```

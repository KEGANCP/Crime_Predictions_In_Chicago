```python
#import dependencies
import pandas as pd
import datetime as dt
```

```python
#load raw csv file

weather_df = pd.read_csv('./weather_data_2010_present.csv')
weather_df
```

```python
#check all column header names
weather_df.columns
```

```python
#rename columns
weather_df.rename({'STATION': 'Station',
                   'LATITUDE': 'Latitude',
                   'LONGITUDE': 'Longitude',
                   'ELEVATION': 'Elevation',
                   'DATE': 'Date',
                   'AWND': 'Average Wind Speed',
                   'FMTM': 'Time of Fastest Wind',
                   'PGTM': 'Peak Gust Time',
                   'TAVG': 'Average Temperature',
                   'TMAX': 'Maximum Temperature',
                   'TMIN': 'Minimum Temperature',
                   'WDF2': 'Direction Fastest 2-Minute Wind',
                   'WDF5': 'Direction Fastest 5-Minute Wind',
                   'WSF2': 'Fastest 2-Minute Wind Speed',
                   'WSF5': 'Fastest 2-Minute Wind Speed',
                   'WT01': 'Fog, Ice, Freezing Fog',
                   'WT02': 'Heavy Fog or Heavy Freezing Fog',
                   'WT03': 'Thunder',
                   'WT04': 'Ice Pellets, Sleet or Small Hail',
                   'WT05': 'Hail',
                   'WT06': 'Glaze or Rime',
                   'WT07': 'Dust, Volcanic Ash or Blowing Dust',
                   'WT08': 'Smoke or Haze',
                   'WT09': 'Blowing or Drifting Snow',
                   'WT11': 'High or Damaging Winds',
                   'WT13': 'Mist',
                   'WT14': 'Drizzle',
                   'WT15': 'Freezing Drizzle',
                   'WT16': 'Rain',
                   'WT17': 'Freezing Rain',
                   'WT18': 'Snow, Snow Pellets or Ice Crystals',
                   'WT19': 'Unknown Precipitation',
                   'WT21': 'Ground Fog',
                   'WT22': 'Ice fog or Freezing Fog'
                  }, axis='columns', inplace=True)
weather_df
```

```python
#set date column to datetime format
weather_df['Date'] = pd.to_datetime(weather_df['Date']).dt.date
weather_df
```

```python
#export to csv
output_data_file = "clean_weather_data.csv"
weather_df.to_csv(output_data_file, index=False)
```

```python
#import dependencies
import pandas as pd
import datetime as dt
```

```python
#load raw csv file

weather_df = pd.read_csv('./weather_data_2010_present.csv')
weather_df
```

```python
#check all column header names
weather_df.columns
```

```python
#rename columns
weather_df.rename({'STATION': 'Station',
                   'LATITUDE': 'Latitude',
                   'LONGITUDE': 'Longitude',
                   'ELEVATION': 'Elevation',
                   'DATE': 'Date',
                   'AWND': 'Average_Wind_Speed',
                   'FMTM': 'Time_of_Fastest_Wind',
                   'PGTM': 'Peak_Gust_Time',
                   'TAVG': 'Average_Temperature',
                   'TMAX': 'Maximum_Temperature',
                   'TMIN': 'Minimum_Temperature',
                   'WDF2': 'Direction_Fastest_2_Minute_Wind',
                   'WDF5': 'Direction_Fastest_5_Minute_Wind',
                   'WSF2': 'Fastest_2_Minute_Wind_Speed',
                   'WSF5': 'Fastest_5_Minute_Wind_Speed',
                   'WT01': 'Fog_Ice_Freezing_Fog',
                   'WT02': 'Heavy_Fog_or_Heavy_Freezing_Fog',
                   'WT03': 'Thunder',
                   'WT04': 'Ice_Pellets_Sleet_or_Small_Hail',
                   'WT05': 'Hail',
                   'WT06': 'Glaze_or_Rime',
                   'WT07': 'Dust_Volcanic_Ash_or_Blowing_Dust',
                   'WT08': 'Smoke_or_Haze',
                   'WT09': 'Blowing_or_Drifting_Snow',
                   'WT11': 'High_or_Damaging_Winds',
                   'WT13': 'Mist',
                   'WT14': 'Drizzle',
                   'WT15': 'Freezing_Drizzle',
                   'WT16': 'Rain',
                   'WT17': 'Freezing_Rain',
                   'WT18': 'Snow_Snow_Pellets_or_Ice Crystals',
                   'WT19': 'Unknown_Precipitation',
                   'WT21': 'Ground_Fog',
                   'WT22': 'Ice_fog_or_Freezing_Fog'
                  }, axis='columns', inplace=True)
weather_df
```

```python
#set date column to datetime format
weather_df['Date'] = pd.to_datetime(weather_df['Date']).dt.date
weather_df
```

```python
weather_df.count()
```

```python
#Drop unneeded columns
weather_df = weather_df.drop(columns=[
    'Station',
    'Latitude',
    'Longitude',
    'Elevation',
    'Time_of_Fastest_Wind',
    'Peak_Gust_Time',
    'Direction_Fastest_2_Minute_Wind',
    'Direction_Fastest_5_Minute_Wind',
    'Fastest_2_Minute_Wind_Speed',
    'Fastest_5_Minute_Wind_Speed',
    'Ice_Pellets_Sleet_or_Small_Hail',
    'Glaze_or_Rime',
    'Dust_Volcanic_Ash_or_Blowing_Dust',
    'Blowing_or_Drifting_Snow',
    'High_or_Damaging_Winds',
    'Drizzle',
    'Freezing_Drizzle',
    'Unknown_Precipitation',
    'Ground_Fog',
    'Ice_fog_or_Freezing_Fog'   
])
weather_df.head()
```

```python
#export to csv
output_data_file = "clean_weather_data.csv"
weather_df.to_csv(output_data_file, index=False)
```

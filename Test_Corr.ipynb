```python
import pandas as pd
import numpy as np
from path import Path
```

```python
# Loading crime data
file_path = Path("Resources/sample_crime_data.csv")
df = pd.read_csv(file_path)
df = df.set_index("ID")
print(df.shape)
df.head()
```

```python
df['Violence Status'].value_counts()
```

```python
df = df[(df['Date'] >= '2015-01-01')]
print(df.shape)
df.head()
```

```python
# Loading weather data
file_path = Path("Resources/clean_weather_data.csv")
weather_df = pd.read_csv(file_path)
#df = df.set_index("ID")
print(df.shape)
weather_df.head()
```

```python
weather_df = weather_df[(weather_df['Date'] >= '2015-01-01')]
print(weather_df.shape)
weather_df.head()
```

```python
merged_df = pd.merge(df, weather_df, how='left', on='Date')
```

```python
print(merged_df.shape)
merged_df.head()
```

```python
df = merged_df.replace(np.nan,0)
df.head()
```

```python
df = df.drop(columns=['ID.1', 'Block', 'IUCR','FBI Code','Ward', 'District','Heavy_Fog_or_Heavy_Freezing_Fog', 'Mist', 'Freezing_Rain', 'Thunder',
                                        'Snow_Snow_Pellets_or_Ice Crystals', 'Hail', 'Rain',
                                       'Maximum_Temperature', 'Minimum_Temperature'])
```

```python
df.nunique()
```

```python
df.corr()
```

```python
ml_df = df.drop(columns=['Arrest','Domestic','Year','Latitude', 'Longitude'])
print(ml_df.shape)
ml_df.head()
```

```python
ml_df.corr()
```

```python
prim_df = df[['Primary Type', 'Violence Status', 'Community Area', 'Average_Wind_Speed', 'Average_Temperature']]
prim_df = pd.get_dummies(prim_df)
print(prim_df.shape)
prim_df.head()
```
